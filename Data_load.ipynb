```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#create database connection
engine = create_engine(
    "postgresql://postgres:root@localhost:5432/ecommerce_intelligence"
)
```

```python
#load csv files
fact_sales = pd.read_csv("cleaned_ecommerce_sales_2025.csv")
monthly_kpi = pd.read_csv("monthly_kpi.csv")
category_kpi = pd.read_csv("category_kpi.csv")
risk_table = pd.read_csv("risk_table.csv")
customer_kpi = pd.read_csv("customer_kpi.csv")
```

```python
#Insert data into tables
#use if_exists='append'(because the table already exist)
#fact sales
fact_sales.to_sql(
    "fact_sales",
     engine,
     if_exists="append",
     index=False
)
```

```python
monthly_kpi.to_sql(
    "monthly_kpi",
    engine,
    if_exists="append",
    index=False
)
     
     
```

```python
category_kpi.to_sql(
    "category_kpi",
    engine,
    if_exists="append",
    index=False
)
```

```python
risk_table.to_sql(
    "risk_table",
    engine,
    if_exists="append",
    index=False
)
```

```python
customer_kpi.to_sql(
    "customer_kpi",
    engine,
    if_exists="append",
    index=False
)
```
