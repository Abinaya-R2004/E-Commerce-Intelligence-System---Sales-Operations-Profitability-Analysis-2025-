# Category KPI Table(Product Intelligence)
#Purpose : Category profitability & risk

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\abina\cleaned_ecommerce_sales_2025.csv")
df.head()
```

```python
category_kpi = (
    df.groupby('product_category')
    .agg(
        total_orders=('order_id', 'count'),
        total_revenue=('net_order_value', 'sum'),
        avg_discount=('discount_percent', 'mean'),
        avg_rating=('customer_rating', 'mean'),
        return_rate=('is_returned', 'mean')
    )
    .reset_index()
)

category_kpi
```

```python
#To display rounded number with decimal value as 2
pd.options.display.float_format = '{:,.2f}'.format
category_kpi
```

```python
category_kpi.to_csv("category_kpi.csv",index=False)
```
