# Risk Table(Core of Intelligence System)
#Purpose:Early warning system

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\abina\cleaned_ecommerce_sales_2025.csv")
```

```python
risk_df = df.copy()
```

```python
risk_df['delivery_risk'] = np.where(risk_df['delivery_days'] > 7, 1, 0)
risk_df['return_risk'] = np.where(risk_df['is_returned'] == 1, 1, 0)
risk_df['rating_risk'] = np.where(risk_df['customer_rating'] < 3, 1, 0)
risk_df['discount_risk']=np.where(risk_df['discount_percent']>30,1,0)
```

```python
#Risk Score
risk_df['risk_score'] = (
    risk_df['delivery_risk'] +
    risk_df['return_risk'] +
    risk_df['rating_risk'] +
    risk_df['discount_risk']
)
```

```python
#Risk Level
risk_df['risk_level'] = pd.cut(
    risk_df['risk_score'],
    bins=[-1, 1, 2, 4],
    labels=['LOW','MEDIUM','HIGH']
)
```

```python
#Final Risk Table
risk_table = risk_df[[
    'order_id',
    'customer_id',
    'product_category',
    'region',
    'net_order_value',
    'risk_score',
    'risk_level'
]]

risk_table
```

```python
risk_table.to_csv("risk_table.csv",index=False)
```
