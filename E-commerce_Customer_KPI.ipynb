#Customer KPI Table(Customer Intelligence)
##Purpose:Loyalty & churn risk

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\abina\cleaned_ecommerce_sales_2025.csv")
```

```python
customer_kpi = (
    df.groupby('customer_id')
    .agg(
        total_orders=('order_id', 'count'),
        total_spend=('net_order_value', 'sum'),
        avg_rating=('customer_rating', 'mean'),
        return_rate=('is_returned', 'mean'),
        avg_delivery_days=('delivery_days', 'mean')
    )
    .reset_index()
)
```

```python
# Customer Segmentation
customer_kpi['customer_segment'] = pd.cut(
    customer_kpi['total_spend'],
    bins=[-1, 500, 2000, customer_kpi['total_spend'].max()],
    labels=['Low Value', 'Medium Value', 'High Value']
)

customer_kpi
```

```python
customer_kpi.to_csv("customer_kpi.csv",index=False)
```
