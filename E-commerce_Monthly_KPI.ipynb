#Monthly KPI Table(Business Performance)
##Purpose:Revenue trend,returns,delivery health

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\abina\cleaned_ecommerce_sales_2025.csv")
df.head()
```

```python
# Data Preparation
# Convert date
df['order_date'] = pd.to_datetime(df['order_date'])

# Create calculated columns
df['order_value'] = df['product_price'] * df['quantity']
df['net_order_value'] = df['order_value'] * (1 - df['discount_percent'] / 100)

# Extract month
df['month'] = df['order_date'].dt.to_period('M')
```

```python
monthly_kpi = (
    df.groupby('month')
    .agg(
        total_orders=('order_id', 'count'),
        total_revenue=('net_order_value', 'sum'),
        avg_order_value=('net_order_value', 'mean'),
        return_rate=('is_returned', 'mean'),
        avg_delivery_days=('delivery_days', 'mean')
    )
    .reset_index()
)

monthly_kpi
```

```python
monthly_kpi.to_csv("monthly_kpi.csv",index=False)
```
